```
from numpy import random
import numpy as np
import matplotlib.pyplot as plt
```

I am going to use estimating Pi using the Monte Carlo Method.

```
# n is the number of points
def fill_square(n = 1000):
  circle_x = []
  circle_y = []
  square_x = []
  square_y = []

  for i in range(n):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if (x**2 + y**2 <= 1):
      circle_x.append(x)
      circle_y.append(y)
    else:
      square_x.append(x)
      square_y.append(y)

  return [n, circle_x, circle_y, square_x, square_y]
```

```
def estimate_pi(n, circle):
  return 4 * len(circle) / float(n)
```

```
board_params = fill_square(1000000)
pi_est = estimate_pi(board_params[0], board_params[1])
```

```
print('The estimated value of pi is: ' + str(pi_est))
print('The real value of pi is: ' + str(np.pi))
```

```
# Let's plot the board
plt.figure(figsize = (6, 6))
plt.plot(board_params[1], board_params[2], 'r.')
plt.plot(board_params[3], board_params[4], 'b.')
plt.grid()
```

Now, let's run the estimation function multiple times and take averages of results. This will give us better estimations with less simultaniously placed points on the board.

```
def pi_est_repeated(iterations, n):
  results = []
  avg_pi_values = []
  for i in range(iterations):
    board_params = fill_square(n)
    results.append(estimate_pi(board_params[0], board_params[1]))
    avg_pi = sum(results) / len(results)
    avg_pi_values.append(avg_pi)
  
  return avg_pi_values
```

```
avg_pi_values = pi_est_repeated(1000, 1000)
```

```
avg_pi_errors = [abs(np.pi - pi) for pi in avg_pi_values]
```

```
# Here we can see the error over iterations 
plt.plot(avg_pi_errors)
```

```
print('The final estimated value of pi is: ' + str(avg_pi_values[len(avg_pi_values) - 1]))
print('The real value of pi is: ' + str(np.pi))
```

We can see that error decreases with each iteration and the final value of pi is quite close to the actual value.
