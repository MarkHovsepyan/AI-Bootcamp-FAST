Task 1: If the sum of 2 numbers is even return **False**, else return **True**.

```python
def my_operation(x,y):
    if (x + y) % 2 == 0:
      return False
    else:
      return True
```

```python
assert(my_operation(2,2) == False)
assert(my_operation(0,0) == False)
assert(my_operation(-2,86) == False)
assert(my_operation(-2,-155) == True)
assert(my_operation(1,8) == True)
assert(my_operation(78,-78) == False)
```

Task 2: Find numbers from 1 to 1000 which are multiples of 13, then put them in the list.

```python
squares = [i for i in range(1, 1000) if i % 13 == 0]
```

```python
assert(squares[0] == 13)
assert(squares[12] == 169)
assert(squares[-1] == 988)
assert(squares[67] == 884)
assert(squares[38] == 507)
```

Task 3: Define a function which calculates the how many 0-s are in a list.

```python
def calc_zeros(x):
    zeros = [element for element in x if element == 0]
    return len(zeros)
```

```python
assert(calc_zeros([1,2,4,241,241,5,3,7,23,0]) == 1)
assert(calc_zeros([0,0,0]) == 3)
assert(calc_zeros([1,0,1,0,1,0,0,1,0,10,1]) == 5)
```

Task 4: Define a function which calculates the sum of 2 arguments if the sum is less than 100, else it must return 0.

```python
def conditional_sum(a,b):
    return a + b if a + b < 100 else 0
```

```python
assert(conditional_sum(1,1) == 2)
assert(conditional_sum(1,100) == 0)
assert(conditional_sum(0,0) == 0)
assert(conditional_sum(-1,-61) == -62)
assert(conditional_sum(-101,1000) == 0)
```

Task 5: Write a for loop which will add all string elements in list_to_be_reduced to reduced_string.

```python
list_to_be_reduced = ['He',0,'ll',89,'o, ',444,'my',' ','na',89,'me i','s A',798,'rth',0,'ur.']
reduced_string = ''

for e in list_to_be_reduced:
  if isinstance(e, str):
    reduced_string += e
```

```python
assert(reduced_string == 'Hello, my name is Arthur.')
```

Task 6: Write a function which will convert seconds to (hour,minute,seconds).

```python
def clock(s):
  m, s = divmod(s, 60)
  h, m = divmod(m, 60)
  d, h = divmod(h, 24)
  return (h, m, s)
```

```python
assert(clock(658400) == (14, 53, 20))
assert(clock(12312412) == (12, 6, 52))
assert(clock(789) == (0, 13, 9))
```

Task 7: Calculate elementwise sums of 2 lists and return the sums as a list.

```python
def list_sums(a,b):
    return [sum(x) for x in zip(a, b)]
```

```python
assert(list_sums([1,1,1],[2,2,2]) == [3,3,3])
assert(list_sums([1,21,1,234],[232,2,78,3]) == [233, 23, 79, 237])
assert(list_sums([1,21,1432,234,434,232],[232,2,78,3,432,-3243290]) == [233, 23, 1510, 237, 866, -3243058])
```

Task 9: Create Fibonacci sequence up to 10000  in a list.

```python
a = 0
b = 1
fib_list = [a, b]

for i in range(2, 10000):
  fib_list.append(fib_list[i - 1] + fib_list[i - 2])
```

```python
assert(fib_list[0] == 0)
assert(fib_list[9] == 34)
assert(fib_list[12] == 144)
assert(fib_list[17] == 1597)
assert(fib_list[20] == 6765)
```
