### Error Handling

```python
# you can handle errors using try, except syntax
x = 0
try:
    y = 1/x
    print('divided succesfully')
except:
    print('divided by zero')
    
x = 2
try:
    y = 1/x
    print('divided succesfully')
except:
    print('divided by zero')
```

```python
# to handle specific error as ZeroDivisionError you can use the following syntax
x = 0
try:
    y = 1/x
    print('divided succesfully')
except ZeroDivisionError:
    print('divided by zero')
```

```python
# Lets put another type of error in try clause
x = 0
try:
    y = x + 'a'
    print('divided succesfully')
except ZeroDivisionError:
    print('divided by zero')
```

```python
# You can add multipe errors in the same except clause or write separate except clause for type of error
x = 0
try:
    y = x + 'a'
    print('divided succesfully')
except (ZeroDivisionError,TypeError) as e:
    print(e)    

x = 0
try:
    y = x + 'a'
    print('divided succesfully')
except ZeroDivisionError:
    print('divided by zero')
except TypeError as e:
    print(e)    
```

There are else and finally clause. Else clause is executed if no errors occur. This is useful to keep your code understandable and separating parts of code which raise errors and run without errors. Finally clause will execute regardless errors were met or not.

```python
x = 1
try:
    y = 1/x
except ZeroDivisionError:
    print('divided by zero')
else:
    print('divided succesfully')
finally:
    print('end of script')
```

```python
x = 5
try:
    y = 1/x
except ZeroDivisionError:
    print('divided by zero')
else:
    print('divided succesfully')
finally:
    print('end of script')
```

```python
import numpy as np
```

Let's make a general class for guns

```python
class gun:
    # The first argument in methods is referred to the class itself
    # The convention is to write 'self'
    def fire(self, distance, time, bullets_per_second):
        """Caculates average accuracy for given parameters"""
        accuracy = 1/np.log(distance)
        return time*bullets_per_second*accuracy 
```

```python
# create new object isntance
ak74 = gun()
```

```python
# apply fire method
ak74.fire(300,10,25)
```

```python
class gun:

    # this is the constructor of the class
    # it is called when new object isntance is created
    def __init__(self, name):
        self.name = name
        
    def fire(self, distance, time, bullets_per_second):
        """Caculates average accuracy for given parameters"""
        accuracy = 1/np.log(distance)
        return time*bullets_per_second*accuracy 
```

```python
ak74 = gun()
```

```python
class gun:

    # now another variable tied to the gun type is added
    def __init__(self, name, bullets_per_second):
        self.name = name
        self.bullets_per_second = bullets_per_second
        
    def fire(self, distance, time):
        """Caculates the average number of hit bullets for given parameters"""
        accuracy = 1/np.log(distance)
        return time*self.bullets_per_second*accuracy # notice the 'self' 
```

```python
ak74 = gun('ak74',25)
```

```python
ak74.fire(300,10)
```

### Inheritance

The inheritance is written as an _'argument'_ to class function.

```python
class customized_gun(gun):
    
    # *args means any number of argumnents may be passed
    # **kwargs means any number of keyword arguments may be passed
    # You may notate as *, ** or *a, **kw, but it is encouraged to use the approach below
    def __init__(self, color, *args, **kwargs):
        self.color = color
        # super calls the function from parent class
        super().__init__(*args,**kwargs)
```

```python
my_gun = customized_gun('green', name = 'ak_mod',bullets_per_second = 30)
my_gun.fire(300,10)
```

```python
my_gun = customized_gun('green','ak_mod',30)
my_gun.fire(300,10)
```

You can overwrite the parent method.

```python
class customized_gun(gun):
    
    def __init__(self, color, *args, **kwargs):
        self.color = color
        super().__init__(*args,**kwargs)
        
    def fire(self, distance, time):
        """Caculates the average number of hit bullets for given parameters"""
        accuracy = 1/np.log(distance/5)
        return time*self.bullets_per_second*accuracy # notice the 'self' 
```

```python
class customized_gun(gun):
    
    def __init__(self, color, *args, **kwargs):
        self.color = color
        super().__init__(*args,**kwargs)
        
    def fire(self, distance, time):
        return super().fire(distance,time)
    
```

```python
my_gun = customized_gun('green','ak_mod',25)
print(my_gun.name,my_gun.fire(300,10))
my_gun = gun('ak74',25)
print(my_gun.name,my_gun.fire(300,10))
```

As you can see **fire** method returns different results for different classes, altough it has the same arguments.

For more info on classes follow https://docs.python.org/3/tutorial/classes.html.
