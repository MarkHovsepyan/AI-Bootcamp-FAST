# Supplement for Linear Regression

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.regression.quantile_regression import QuantReg
from sklearn.metrics import r2_score
```

### Difference between MAD and MSE

```
# Generate test data
X = np.array([i for i in range(20)]).reshape(-1, 1)
Y = np.array([np.random.random() for i in range(19)]+[50])
```

```
# Plot
plt.figure(figsize = (16,8))
plt.plot(X,Y,'bo')
plt.xlabel('X',fontsize = 18)
plt.ylabel('Y',fontsize = 18)
plt.show()
```

```
mod = LinearRegression()
res = mod.fit(X,Y)

# Plot
plt.figure(figsize = (16,8))
plt.plot(X,Y,'bo')
plt.plot(res.predict(X))
plt.xlabel('X',fontsize = 18)
plt.ylabel('Y',fontsize = 18)
plt.title('Regression with MSE', fontsize = 24)
plt.show()
```

```
mod = QuantReg(Y,X)
res = mod.fit(q=.5)

# Plot
plt.figure(figsize = (16,8))
plt.plot(X,Y,'bo')
plt.plot(res.predict(X))
plt.xlabel('X',fontsize = 18)
plt.ylabel('Y',fontsize = 18)
plt.title('Regression with MAD', fontsize = 24)
plt.show()
```
