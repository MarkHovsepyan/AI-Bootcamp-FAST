### Creating data with pandas

```python
# creating columns with simple operations in pandas
df['Survived_Age'] = df['Age']*df['Survived']
```

```python
df.head()
```

```python
# for more complicated operations we need to use apply function
df['Age'].apply(lambda x: x**2+x-5)
```

```python
# lambda is equivalent to def, but it is a one liner fo simple functions\
my_func = lambda x: x**3+5*x-4
my_func(1)
```

So apply functions iterates for each value in 'Age' column and applies lambda function to each value.<br>
You can define more complex functions and then use it in apply.

```python
def age_discriminator(age):
    
    """Function which divides age into 3 groups
    
       age: age of a person
       
       return: one of the 3 groups (child, adult, elder)
    """
    
    if age < 18:
        return 'child'
    elif 65 > age >= 18:
        return 'adult'
    else:
        return 'elder'
```

```python
df['Age'].apply(age_discriminator)
```

```python
# You can use fillna to assign values to NA 
df = pd.read_csv(path+'train.csv')
print('Age mean:',df['Age'].mean())
df['Age'].fillna(df['Age'].mean())
```

```python
# To filter values you can use the following syntax
df[df['Age']>65]
```

```python
# multiple filters
df[(df['Age']>65)&(df['Pclass'] == 1)]
```

Suppose you need to get mean age for each Passenger class. A simple approach would be to create separate filters for each 'Pclass', then calculate mean of 'Age'. In this case **groupby** function comes in handy.

```python
groups = df.groupby('Pclass')
```

As you can see each variable is separated by 'Pclass', then an average for each is returned.

```python
groups.mean()
```

```python
# standard deviation for each variable in each group.
groups.std()
```

### Manipulating dataframes

```python
data1 = [[1,2,3],[1,3,5]]
df1 = pd.DataFrame(data1,columns = ['A','B','C'])
data2 = [[4,2,7],[5,8,9]]
df2 = pd.DataFrame(data2,columns = ['A','B','D'])
```

```python
df1
```

```python
df2
```

```python
# append
df1.append(df2)
```

```python
# concat
pd.concat([df1,df2], axis = 1)
```

```python
# join
df1.join(df2,lsuffix = '0')
```

```python
# merge
pd.merge(df1,df2,on = ['B'], how='outer')
```

For more detailed guide on data joining, merging, etc. follow the link https://pandas.pydata.org/pandas-docs/stable/merging.html.

# Python - Visualization

This part of collab notebook is an intorduction to Python visualization tools like **matplotlib, seaborn**.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [i for i in range(100)]
y = np.random.random(100)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# giving names to axes 
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# adding title 
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random')
plt.show()
```

```python
# resizing plot
plt.figure(figsize = (16,8))
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Random')
plt.show()
```

```python
# histograms
x = np.random.normal(size = 10000)
plt.figure(figsize = (16,8))
plt.hist(x, bins = 80)
plt.title('Normal (0,1)', fontsize = 18)
plt.show()
```

```python
plt.figure(figsize = (16,8))
x = np.random.normal(loc = 2, size = 1000)
y = np.random.normal(loc = 5, scale = 25, size = 1000)
plt.scatter(x,y)
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.show()
```

```python
# stacking plots
plt.figure(figsize = (16,8))
x1 = np.random.normal(loc = 2, size = 1000)
y1 = np.random.normal(loc = 5, scale = 25, size = 1000)
x2 = np.random.normal(loc = 5, scale = 0.5,size = 1000)
y2 = np.random.normal(loc = 10, scale = 40, size = 1000)
plt.scatter(x1,y1, c = 'b')
plt.scatter(x2,y2, c = 'r')
plt.xlabel('X', fontsize = 12)
plt.ylabel('Y', fontsize = 12)
plt.show()
```

```python
# subplots
fig, axes = plt.subplots(2,2, figsize = (16,8))

axes[0][0].plot(x,y,'bo')
axes[0][0].set_title('scatter')
axes[0][1].hist(x, bins = 40)
axes[0][1].set_title('X histogram')
axes[1][0].hist(y, bins = 40)
axes[1][0].set_title('Y histogram')
# if you don't need a particular subplot
fig.delaxes(axes[1][1])

plt.show()
```

To see many more options in matplotlib follow https://matplotlib.org/3.0.2/gallery/index.html.

### Plotting with seaborn

Seaborn is a high level plotting library which has many useful plotting options to make visual data analysis faster.

```python
import seaborn as sns
```

```python
# plot different plots for defined class (col) on x and y axes separated with style class
sns.relplot(x="Age", y="Fare", col="Survived",
            style="Pclass",
            data=df, height = 8.5, aspect = 0.7)
# height and aspect change the figure size
# You can add hue keyword which will differentiate classes with colors
# You can add size keyword which will differentiate classes with size
```

```python
# An example of using the other differentiating keywords
# Try not to use many is it is becoming complicated
sns.relplot(x="Age", y="Fare", col="Survived",
            style="Pclass", size = 'Sex', hue = 'SibSp',
            data=df, height = 8.5, aspect = 0.7)
plt.show()
```

```python
# This will plot pair by pair variables separated by given class
sns.pairplot(data=df, hue="Survived")
plt.show()
```

```python
# You can plot correlation heatmap with sns
corr = df.corr()
plt.figure(figsize = (16,8))
sns.heatmap(corr, annot = True)
plt.show()
```
