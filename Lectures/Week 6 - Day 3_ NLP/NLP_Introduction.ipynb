# Text Preproccesing for modeling

## why and when?

Depending on the task you are solving different preprocessing schemas might be needed. E.g. if you want to do document classification, say spam filtering, in general you don't vcare if a word is in present continous or present simple. What you will care is the overall aggregateted fetaures of the document. However if your task is a NER task, then keeping capital letters, punctuation might be crucial. In general you should identify if your task is an aggregation based or detail based, then decide what kind of preproccesing shceme you need. 

## Text Representation

After doing all the nitty-gritty preprocessing, you will need to have representation for your text data, as neural networks can't just accept text, they need numbers. 

Text is usually represented as vector of words (sometimes characters, phonemes).

```
text
```

```
text.split()
```

The process of dividing text into words (characters, phonemes) is called **tokenization** and the resulting vector consists of **tokens** (words in our case).

Now we need to convert tokens to some numerical representation. The Most simple way would be to consider all unqiue words in our dataset and make a one-hot encoding. Considering that there may be 100,000 and more unqiue words in a given dataset, this is a huge computational drwaback. Besides that, synonyms, antonims and other specific pair of words are orthogonal to each other, as any pair of word. This means that similarity between two words is just ignored. This idea led to Word Embeddings idea. (We will do through lecture on Word Embeddings in upcoming lectures)

## Brief overview of Word embeddings

So the idea of Word Embeddings is to represent each word with a fixed size vector, where vectors are decided from the context of those words.

## NLTK Toolkit

python library nltk is the most comprehensive tool for text preprocessing.

```
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import inflect
import string, unicodedata
```

```
def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words

remove_non_ascii(text.split())
```

```
def to_lowercase(words):
    """Convert all characters to lowercase from list of tokenized words"""
    return [w.lower() for w in words]

to_lowercase(text.split())
```

```
def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words

remove_punctuation(text.split())
```

```
def replace_numbers(words):
    """Replace all integer occurrences in list of tokenized words with textual representation"""
    p = inflect.engine()
    new_words = []
    for word in words:
        if word.isdigit():
            new_word = p.number_to_words(word)
            new_words.append(new_word)
        else:
            new_words.append(word)
    return new_words

replace_numbers(text.split())
```

```
def remove_stopwords(words):
    """Remove stop words from list of tokenized words"""
    return [w for w in words if w not in stopwords.words('english')]

remove_stopwords(text.split())
```

```
def stem_words(words):
    """Stem words in list of tokenized words"""
    stemmer = PorterStemmer()
    return [stemmer.stem(w) for w in words]

stem_words(text.split())
```

```
def lemmatize_verbs(words):
    """Lemmatize verbs in list of tokenized words"""
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(w, pos='v') for w in words]

lemmatize_verbs(text.split()) 
```
