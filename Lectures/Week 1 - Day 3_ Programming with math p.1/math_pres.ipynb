```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns; sns.set()
```

```python
# f(x) = x^2 function tangents
x = [i/100 for i in range(-100,100)]
y = [i**2 for i in x]

def tangent_y(p):
     return [2*p*i-p**2 for i in x]

plt.figure(figsize = (8,7*2/3))
plt.plot(x,y, label = 'f(x)',color = 'b')
for k,c in zip([0.8,0.4,0],['r','g','black']):
    start = int(50+100*k)
    plt.plot(x[start:],tangent_y(k)[start:], label = 'tangent at '+str(k), color = c)
    plt.plot(k,k**2, 'o', color = c)
plt.legend()
plt.savefig('x^2_tangents.png', bbox_inches = 'tight')
plt.show()
```

```python
# f(x) = x^2 function tangents
x = [i/100 for i in range(-100,100)]
y = [i**2 for i in x]

def line(s,e):
    if e!= s:
        slope = (e**2-s**2)/(e-s)
        intercept = s**2-slope*s
    else:
        slope = 2*s
        intercept = s**2-slope*s
    return [slope*i+intercept for i in x]

plt.figure(figsize = (10,6))
plt.plot(x,y, label = 'f(x)', color = 'b')
for k,c in zip([0.8,0.4,0,-0.4],['r','g','black','y']):
    end = int(170+30*k)
    start = int(20+30*k)
    plt.plot(x[start:end],line(-0.4,k)[start:end], label = 'line from -0.4 to '+str(k), color = c)
    plt.plot(k,k**2, 'o', color = c)
plt.legend()
plt.savefig('x^2_lines.png', bbox_inches = 'tight')
plt.show()
```

```python
def test_func(i):
    return i**4+2*i**3-19*i**2+6*i
```

```python
x = [i/100 for i in range(-600,400)]
y = [i**4+2*i**3-19*i**2+6*i for i in x]

plt.figure(figsize = (6,4))
plt.plot(x,y,label = 'x^4+2x^3-19x^2+6x')
plt.plot(2.32,test_func(2.32),'ro')
plt.plot(-3.98,test_func(-3.98),'ro')
plt.plot(0.16,test_func(0.16),color = 'black', marker = 'o')
plt.legend()
plt.savefig('multipleextremum.png', bbox_inches = 'tight')
plt.show()
```

```python
# for contour plots
def f(x, y):
    return x**2+y**2

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize = (6,4))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 100, cmap='RdGy')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig('contour3d_simple.png', bbox_inches = 'tight')
plt.show()

plt.figure(figsize = (6,6))
plt.contour(X, Y, Z, 30, cmap='RdGy')
plt.savefig('contour_simple.png', bbox_inches = 'tight')
plt.show()
```

```python
# for contour plots
def f(x, y):
    return np.cos(2*y)*np.cos(x*y)
```

```python
x = np.linspace(2, 5, 100)
y = np.linspace(2, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize = (6,4))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='RdGy')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.savefig('contour3d.png', bbox_inches = 'tight')
plt.show()
```
